```python
# Función para eliminar caracteres invisibles (como espacios, saltos de línea, tabulaciones, etc.)
def eliminar_caracteres_invisibles(valor):
    if isinstance(valor, str):
        # Eliminar caracteres no visibles (como espacios no rompibles, tabulaciones, saltos de línea, etc.)
        return ''.join(c for c in valor if c.isprintable())
    return valor  # Si no es una cadena, no hacer nada

# Aplicar la función a todo el DataFrame
df_unido3 = df_unido3.applymap(eliminar_caracteres_invisibles)
```
